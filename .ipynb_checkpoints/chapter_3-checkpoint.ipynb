***재표본 추출 for 가설 검정***

```python
# 웹페이지 A와 B의 세션시간 차이
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

session_times = pd.read_csv('/Users/wonbinchoi/current_studying/Practical_Statistics_for_Data_Scientists/data/web_page_data.csv')
```

```python
# 페이지 B가 방문객들을 더 오래 붙잡은 것을 알 수 있음. 
ax = session_times.boxplot(by='Page', column='Time')
ax.set_xlabel('')
ax.set_ylabel('Time (in seconds)')
plt.tight_layout()
plt.show()
```

```python
# 순열 검정을 통해 위의 차이가 우연에 의한 것인지 
# 아니면 통계적으로 중요한 것인지를 판단하도록 도와줌.
import random
import warnings
warnings.filterwarnings('ignore')

# 순열검정 (재표본추출) 함수 ==> 임의 순열 검정이라고 부름. 
# 비복원
def perm_func(x, nA, nB):
    n = nA + nB
    idx_B = set(random.sample(range(n), nB)) # 순차적으로 정렬된 range(n)에서 nB만큼 무작위로 뽑음. 
    idx_A = set(range(n)) - idx_B
    return x.loc[idx_B].mean() - x.loc[idx_A].mean()
nA, nB = 21, 15
perm_diffs = [perm_func(session_times.Time, nA, nB) for _ in range(1000)]
perm_diffs
```

```python
mean_a = session_times[session_times.Page == 'Page A'].Time.mean()
mean_b = session_times[session_times.Page == 'Page B'].Time.mean()

fig,ax = plt.subplots(figsize=(10,10))
ax.hist(perm_diffs, bins=11, rwidth=.9)
ax.axvline(x=mean_b-mean_a, color='black', lw=2)
# ax.text(50,190,'Observed\ndiference')
ax.set_xlabel('Session time differences (in seconds)')
ax.set_ylabel('Frequency')
plt.show()
```

```python
# 결론
# 무작위 순열로 구한 평균 세션 시간의 차이가 가끔 실제 관찰된 세션 시간의 차이 (수직선)을 넘어감을 알 수 있음.
# 이는 페이지 A와 페이지 B 사이의 세션 시간의 차이가 확률분포의 범위 내에 있음을 의미하고, 
# 따라서 통계적으로 유의하지 않음. 

## A,B 테스팅 때 유용하게 사용할 수 있음. 
# 전환율에도 쓰일 수 있음 (여기서는 생략)
```

***3.5 t 검정*** <<< 사용

```python
# t 검정을 위한 scipy.stats.ttest_ind
# 다양한 다른 요인에 대해 걱정하지 않고 p_value를 구할 때 쓸 수 있음
# 즉, 순열 검정 대신 t 검정을 이용하는 것 
from scipy import stats

res = stats.ttest_ind(
    session_times[session_times.Page == 'Page A'].Time,
    session_times[session_times.Page == 'Page B'].Time,
    equal_var=False
)
print('p-value for single sided test : {0: .4f}'.format(res.pvalue / 2))
```

***3.8 분산분석***

```python
four_sessions = pd.read_csv('/Users/wonbinchoi/current_studying/Practical_Statistics_for_Data_Scientists/data/four_sessions.csv')
```

```python
# A/B 테스트가 아닌 A/B/C/D 테스트가 가능하도록 도와줌. 
# 웹테스트에서는 일부 방문자를 어떤 커다란 모집단에서 무작위로 선택하는 식의, 전통적인 임의표본추출 설계 완전 구현 X
# 방문자는 오는 대로 바로 대상이 되고 / 요일, 계절, 인터넷 환경, 사용하는 장치 등에 따라 체계적으로 다를 수 있음.
# 실험 결과를 검토할 때 이러한 요소들을 잠재적 편향의 요인으로 고려하기. 

observed_variance = four_sessions.groupby('Page').mean().var()[0] # 네 개의 페이지 평균의 분산
print('observed means : ', four_sessions.groupby('Page').mean().values.ravel())
print('observed variance : ', observed_variance)

def perm_test(df):
    df = df.copy()
    df['Time'] = np.random.permutation(df['Time'].values) # 재표본추출로 무작위로 섞음 (random sampling)
                                                          # 그리고 다시 df['Time']에 넣음
    return df.groupby('Page').mean().var()[0]
    
perm_variance = [perm_test(four_sessions) for _ in range(3000)] # 해당 과정을 반복
print('Pr(Prob)', np.mean([var > observed_variance for var in perm_variance])) 

# ==> 4 페이지 간의 차이가 우연히 발생할 수 있다는 결론을 내림. 
```

***3.8.1 F 통계량*** <<< 사용

```python
# ANOVA 구현 
# 위에서는 순열 검정을 통해 평균을 비교했다면, 
# 아래에서는 A/B 테스트에서 t 검정을 대신 사용한 것 처럼 F 통계량 사용

import statsmodels.formula.api as smf
import statsmodels.api as sm

model = smf.ols('Time ~ Page', data=four_sessions).fit()
anova_table = sm.stats.anova_lm(model)
anova_table
```

***3.11 검정력과 표본크기***

```python
# ex) 웹 테스트에를 수행할 경우 실행 시간(즉, 처리당 얼마나 많은 노출이 필요할까?)는 어떻게 결정?
#     원하는 달성 목표에 따라 조절. 

# 검정력 혹은 필요한 표본크기의 계산과 관련한 다음 4가지 중요한 요소들이 있음.
# 1. 표본크기
# 2. 탐지하고자 하는 효과크기
# 3. 가설검정을 위한 유의수준
# 4. 검정력 
```

```python
import statsmodels.api as sm
import warnings 
warnings.filterwarnings('ignore')

effect_size = sm.stats.proportion_effectsize(.0121, .011) # 효과크기 계산
analysis = sm.stats.TTestIndPower()
result = analysis.solve_power(effect_size=effect_size, # 계산된 효과크기 
                              alpha=.05, # 검정을 수행할 유의 수준
                              power=.8,  # 검정력 (효과크기를 알아낼 확률) ==> 두 값을 구분할 수 있을 확률
                              alternative='larger' # 단방향 검정
                             )
result
```
